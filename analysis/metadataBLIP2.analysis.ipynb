```python
import os
import pandas as pd

csv_path = 'metadata_blip2.csv'
df = pd.read_csv(csv_path)
```

```python
def emptyRows(df):
    empty_subset = df[df['subset'].isnull() | (df['subset'] == '')]
    empty_genre = df[df['genre'].isnull() | (df['genre'] == '')]
    empty_artist = df[df['artist'].isnull() | (df['artist'] == '')]

    print(f"Filas con 'subset' vacío: {len(empty_subset)}")
    print(f"Filas con 'genre' vacío: {len(empty_genre)}")
    print(f"Filas con 'artist' vacío: {len(empty_artist)}")

    if not empty_subset.empty:
        print("\nEjemplo de filas con 'subset' vacío:")
        display(empty_subset)

    if not empty_genre.empty:
        print("\nEjemplo de filas con 'genre' vacío:")
        display(empty_genre)

    if not empty_artist.empty:
        print("\nEjemplo de filas con 'artist' vacío:")
        display(empty_artist)
        
emptyRows(df)
```

```python
genre_mapping = {
    'Impressionism': 'Impressionism',
    'Realism': 'Realism',
    'Romanticism': 'Romanticism',
    'Expressionism': 'Expressionism',
    'Post_Impressionism': 'Post Impressionism',
    'Symbolism': 'Symbolism',
    'Baroque': 'Baroque',
    'Art_Nouveau_Modern': 'Art Nouveau Modern',
    'Abstract_Expressionism': 'Abstract Expressionism',
    'Northern_Renaissance': 'Northern Renaissance',
    'Naive_Art_Primitivism': 'Naive Art Primitivism',
    'Cubism': 'Cubism',
    'Rococo': 'Rococo',
    'Color_Field_Painting': 'Color Field Painting',
    'Pop_Art': 'Pop Art',
    'Early_Renaissance': 'Early Renaissance',
    'High_Renaissance': 'High Renaissance',
    'Minimalism': 'Minimalism',
    'Mannerism_Late_Renaissance': 'Mannerism Late Renaissance',
    'Ukiyo_e': 'Ukiyo e',
    'Fauvism': 'Fauvism',
    'Pointillism': 'Pointillism',
    'Contemporary_Realism': 'Contemporary Realism',
    'New_Realism': 'New Realism',
    'Synthetic_Cubism': 'Synthetic Cubism',
    'Analytical_Cubism': 'Analytical Cubism',
    'Action_painting': 'Action painting'
}
def extract_genre(file_name):
    folder = file_name.split('/')[0]
    return genre_mapping.get(folder, None)

# Llenar los géneros vacíos
df.loc[df['genre'].isnull() | (df['genre'] == ''), 'genre'] = df['file_name'].apply(extract_genre)
emptyRows(df)
```

```python
import pandas as pd


def extract_artist(file_name):
    parts = file_name.split('/')[1].split('_')
    if len(parts) >= 2:
        if '-' in parts[0]:
            artist = parts[0].replace('-', ' ')
        else:
            artist = parts[0]
        return artist
    else:
        print(f"No valid artist name found in: {file_name}")
        return None

df.loc[df['artist'].isnull() | (df['artist'] == ''), 'artist'] = df['file_name'].apply(extract_artist)

emptyRows(df)
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['genre'], random_state=42)

train_distribution = train_df['genre'].value_counts()
test_distribution = test_df['genre'].value_counts()

print("Distribución de géneros en el conjunto de entrenamiento:")
print(train_distribution)

print("\nDistribución de géneros en el conjunto de prueba:")
print(test_distribution)

train_df['subset'] = 'train'
test_df['subset'] = 'test'

df = pd.concat([train_df, test_df])
```

```python
def extract_description(file_name):
    parts = file_name.split('/')[1].split('_')
    if len(parts) >= 3:
        # El nombre de la pintura está entre el nombre del artista y '_resize1024'
        painting_name = parts[1]
        # Reemplazar los guiones por espacios
        description = painting_name.replace('-', ' ')
        return description
    else:
        print(f"No valid description found in: {file_name}") 
        return None

df['description'] = df['file_name'].apply(extract_description)

empty_description = df[df['description'].isnull() | (df['description'] == '')]

print(f"Filas con 'description' vacío: {len(empty_description)}")

```

```python
import os
import pandas as pd

def changeName(df, incorrect_name, correct_name, image_folder):
    correct_artist_name = correct_name.replace('-', ' ')
    
    df['file_name'] = df['file_name'].str.replace(incorrect_name, correct_name, regex=False)

    df.loc[df['file_name'].str.contains(correct_name, na=False), 'artist'] = correct_artist_name
    
    for root, dirs, files in os.walk(image_folder):
        for file in files:
            if incorrect_name in file:
                old_file_path = os.path.join(root, file)
                new_file_name = file.replace(incorrect_name, correct_name)
                new_file_path = os.path.join(root, new_file_name)
                os.rename(old_file_path, new_file_path)
                # print(f"Renombrado: {old_file_path} a {new_file_path}")
    
    return df

# df = changeName(df, "fтФЬ╨│тФм╨╣lix-del-marle", "felix-del-marle", "imagenes/resizeSD/")

```
